```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/WA_Fn-UseC_-HR-Employee-Attrition.csv'
df = pd.read_csv(file_path)
```

```python
df = df.drop(['EmployeeCount', 'Over18', 'StandardHours'], axis=1)
```

```python
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
for col in categorical_cols:
    print(f"\n{col} unique values:")
    print(df[col].unique())
```

```python
df['Education'] = df['Education'].astype('category')
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].astype('category')
df['JobInvolvement'] = df['JobInvolvement'].astype('category')
df['JobSatisfaction'] = df['JobSatisfaction'].astype('category')
df['PerformanceRating'] = df['PerformanceRating'].astype('category')
df['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].astype('category')
df['WorkLifeBalance'] = df['WorkLifeBalance'].astype('category')
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
numerical_cols
```

```python
for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
df.to_csv('cleaned_employee_attrition.csv', index=False)
```

```python
# 1.	What is the relationship between age and attrition?
df['AgeGroup'] = pd.cut(df['Age'], bins=[18, 30, 40, 50, 60, 70])
age_attrition = df.groupby('AgeGroup')['Attrition'].mean().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x='AgeGroup', y='Attrition', data=age_attrition)
plt.title('Attrition Rate by Age Group')
plt.ylabel('Attrition Rate')
plt.xlabel('Age Group')
plt.show()

```

```python
#2.	How does gender affect attrition rates?
gender_attrition = df.groupby('Gender')['Attrition'].mean().reset_index()

plt.figure(figsize=(8, 5))
ax = sns.barplot(x='Gender', y='Attrition', data=gender_attrition, palette='pastel')
plt.title('Overall Attrition Rate by Gender', fontsize=16)
plt.ylabel('Attrition Rate', fontsize=14)
plt.xlabel('Gender', fontsize=14)
plt.ylim(0, 0.2)
```

```python
# Group by Marital Status and calculate attrition percentage
marital_attrition_py = df.groupby('Gender')['Attrition'].mean() * 100


plt.figure(figsize=(6, 4))
sns.barplot(x=marital_attrition_py.index, y=marital_attrition_py.values, palette="Set3")
plt.ylabel("Attrition Rate (%)")
plt.title("Attrition Rate by Marital Status (Python)")
plt.ylim(0, max(marital_attrition_py.values()) + 5)
plt.tight_layout()
plt.show()

# Return the values as a dictionary
marital_attrition_py.round(2).to_dict()
```

```python
#3.	Does marital status influence the likelihood of attrition?
marital_attrition = df.groupby('MaritalStatus')['Attrition'].mean().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x='MaritalStatus', y='Attrition', data=marital_attrition, palette='pastel')
plt.title('Overall Attrition Rate by Marital Status')
plt.ylabel('Attrition Rate')
plt.xlabel('Marital Status')
plt.show()
```

```python
#4.	Which departments have the highest attrition rates?
dept_attrition_py = df.groupby('Department')['Attrition'].mean() * 100

plt.figure(figsize=(6, 4))
sns.barplot(x=dept_attrition_py.index, y=dept_attrition_py.values, palette="Set2")
plt.ylabel("Attrition Rate (%)")
plt.title("Attrition Rate by Department (Python)")
plt.ylim(0, max(dept_attrition_py.values) + 5)
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()

dept_attrition_py.round(2).to_dict()
```

```python
#5.	How does job satisfaction correlate with attrition?
job_satisfaction_corr = df['JobSatisfaction'].corr(df['Attrition'])
satisfaction_attrition = df.groupby('JobSatisfaction')['Attrition'].mean() * 100

plt.figure(figsize=(6, 4))
sns.barplot(x=satisfaction_attrition.index.astype(str), y=satisfaction_attrition.values, palette="coolwarm")
plt.xlabel("Job Satisfaction Level")
plt.ylabel("Attrition Rate (%)")
plt.title("Attrition Rate by Job Satisfaction Level")
plt.ylim(0, max(satisfaction_attrition.values) + 5)
plt.tight_layout()
plt.show()

job_satisfaction_corr, satisfaction_attrition.round(2).to_dict()

```

```python
#6.	What is the relationship between job level and attrition?
job_level_corr = df['JobLevel'].corr(df['Attrition'])
joblevel_attrition = df.groupby('JobLevel')['Attrition'].mean() * 100

# Plot
plt.figure(figsize=(6, 4))
sns.barplot(x=joblevel_attrition.index.astype(str), y=joblevel_attrition.values, palette="viridis")
plt.xlabel("Job Level")
plt.ylabel("Attrition Rate (%)")
plt.title("Attrition Rate by Job Level")
plt.ylim(0, max(joblevel_attrition.values) + 5)
plt.tight_layout()
plt.show()

job_level_corr, joblevel_attrition.round(2).to_dict()
```

```python
#7.	Is there a correlation between monthly income and attrition?


income_corr = df['MonthlyIncome'].corr(df['Attrition'])

plt.figure(figsize=(7, 4))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome', palette="Set1")
plt.title("Monthly Income by Attrition Status")
plt.ylabel("Monthly Income")
plt.tight_layout()
plt.show()

# Show correlation value
income_corr
```

```python
#8.	Does work-life balance rating influence attrition?
wlb_corr = df['WorkLifeBalance'].corr(df['Attrition'])

wlb_attrition = df.groupby('WorkLifeBalance')['Attrition'].mean() * 100

# Plot
plt.figure(figsize=(6, 4))
sns.barplot(x=wlb_attrition.index.astype(str), y=wlb_attrition.values, palette="pastel")
plt.xlabel("Work-Life Balance Rating")
plt.ylabel("Attrition Rate (%)")
plt.title("Attrition Rate by Work-Life Balance")
plt.ylim(0, max(wlb_attrition.values) + 5)
plt.tight_layout()
plt.show()

# Return correlation and attrition rates
wlb_corr, wlb_attrition.round(2).to_dict()
```

```python
#9.	How does overtime work affect attrition rates?

overtime_attrition = df.groupby('OverTime')['Attrition'].mean() * 100

plt.figure(figsize=(5.5, 4))
sns.barplot(x=overtime_attrition.index, y=overtime_attrition.values, palette="Reds")
plt.ylabel("Attrition Rate (%)")
plt.title("Attrition Rate by Overtime Status")
plt.ylim(0, max(overtime_attrition.values) + 5)
plt.tight_layout()
plt.show()

# Return the values
overtime_attrition.round(2).to_dict()
```

```python
#10.	What is the relationship between years since last promotion and attrition?
promotion_corr = df['YearsSinceLastPromotion'].corr(df['Attrition'])

# Group and calculate attrition rate
promotion_attrition = df.groupby('YearsSinceLastPromotion')['Attrition'].mean() * 100

# Plot
plt.figure(figsize=(7, 4))
sns.lineplot(x=promotion_attrition.index, y=promotion_attrition.values, marker='o', color='purple')
plt.xlabel("Years Since Last Promotion")
plt.ylabel("Attrition Rate (%)")
plt.title("Attrition Rate vs. Years Since Last Promotion")
plt.grid(True)
plt.tight_layout()
plt.show()

# Return correlation and attrition rates
promotion_corr, promotion_attrition.round(2).to_dict()
```

```python
#11.	How does time in current role correlate with attrition?
role_corr = df['YearsInCurrentRole'].corr(df['Attrition'])

role_attrition = df.groupby('YearsInCurrentRole')['Attrition'].mean() * 100

# Plot
plt.figure(figsize=(7, 4))
sns.lineplot(x=role_attrition.index, y=role_attrition.values, marker='o', color='teal')
plt.xlabel("Years in Current Role")
plt.ylabel("Attrition Rate (%)")
plt.title("Attrition Rate vs. Years in Current Role")
plt.grid(True)
plt.tight_layout()
plt.show()

# Return correlation and attrition rates
role_corr, role_attrition.round(2).to_dict()
```

```python
#12.	How does business travel frequency affect attrition?
travel_attrition = df.groupby('BusinessTravel')['Attrition'].mean() * 100

# Plot
plt.figure(figsize=(6.5, 4))
sns.barplot(x=travel_attrition.index, y=travel_attrition.values, palette="Blues")
plt.ylabel("Attrition Rate (%)")
plt.title("Attrition Rate by Business Travel Frequency")
plt.ylim(0, max(travel_attrition.values) + 5)
plt.tight_layout()
plt.show()

# Return attrition rates by travel frequency
travel_attrition.round(2).to_dict()
```

```python
#13.	Does distance from home influence employee retention?
distance_corr = df['DistanceFromHome'].corr(df['Attrition'])

df['DistanceBucket'] = pd.cut(df['DistanceFromHome'], bins=[0, 5, 10, 15, 20, 25, 30], labels=['0-5','6-10','11-15','16-20','21-25','26-30'])
distance_bucket_attrition = df.groupby('DistanceBucket')['Attrition'].mean() * 100

# Plot
plt.figure(figsize=(7, 4))
sns.barplot(x=distance_bucket_attrition.index, y=distance_bucket_attrition.values, palette="Greens")
plt.xlabel("Distance from Home (miles)")
plt.ylabel("Attrition Rate (%)")
plt.title("Attrition Rate vs. Distance from Home")
plt.ylim(0, max(distance_bucket_attrition.values) + 5)
plt.tight_layout()
plt.show()

# Return correlation and attrition rate buckets
distance_corr, distance_bucket_attrition.round(2).to_dict()
```
